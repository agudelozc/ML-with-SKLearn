```python
#Importacion de librerias
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#importacion del dataset
dataset = pd.read_csv('C:\\Users\\CRISTIAN\\Desktop\\IBM Attrition Data.csv',index_col=False)
```

```python
#Encuentre la distribución de edad de los empleados en IBM
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(dataset['Age'],bins=50);
plt.xlabel('Age')
plt.ylabel('Number of Persons')
```

```python
#Explorar la deserción por edad
#Convierto si deserto en 1, si no deserto en 0
def conv(x):
    if x == 'Yes':
        return 1
    elif x =='No':
        return 0
    else:
        return -1
dataset['Attrition'] = dataset['Attrition'].apply(conv)
```

```python
#Changed vblex text for numerics
dataset['Department'].replace('Research & Development',1, inplace=True)
dataset['Department'].replace('Sales',2, inplace=True)
dataset['Department'].replace('Human Resources',3, inplace=True)
#---------------------------------------------------------------------------
```

```python
dataset['EducationField'].replace('Life Sciences',1,inplace=True)
dataset['EducationField'].replace('Medical',2,inplace=True)
dataset['EducationField'].replace('Marketing',3,inplace=True)
dataset['EducationField'].replace('Technical Degree',4,inplace=True)
dataset['EducationField'].replace('Other',5,inplace=True)
dataset['EducationField'].replace('Human Resources',6,inplace=True)
```

```python
dataset['MaritalStatus'].replace('Married',1,inplace=True)
dataset['MaritalStatus'].replace('Single',2,inplace=True)
dataset['MaritalStatus'].replace('Divorced',3,inplace=True)
```

```python
dataset.head(50)
```

```python
X=dataset[['Age','Department','DistanceFromHome','Education','EducationField','EnvironmentSatisfaction','JobSatisfaction','MaritalStatus','MonthlyIncome','NumCompaniesWorked','WorkLifeBalance','YearsAtCompany']]
```

```python
X
```

```python
Y=dataset['Attrition']
```

```python
Y
```

```python
#Libraries for logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
scaler = StandardScaler()
segmentacion = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(segmentacion,Y,test_size=0.4, random_state=42)
```

```python
clf= LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
 print("SCORE : ", clf.score(X_test, y_test))
```

```python
 # Realizamos prediccion con nuestro conjunto de test 
y_pred = clf.predict(X_test)
print(accuracy_score(y_pred, y_test))
```

```python
print (accuracy_score(y_test, y_pred))
print (confusion_matrix(y_test, y_pred))
```
